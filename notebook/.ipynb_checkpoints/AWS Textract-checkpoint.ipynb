```python
import boto3
import botocore
import json
import time
import pandas as pd

from trp.trp2 import TDocument, TDocumentSchema
from trp.t_pipeline import order_blocks_by_geo
from trp import Document
from textractprettyprinter.t_pretty_print import Textract_Pretty_Print, get_string
from textractcaller import get_full_json, Textract_API

# t_doc = TDocumentSchema().load(j)
```

```python
session = boto3.session.Session(
    aws_access_key_id=AWS_ACCESS_KEY_ID,
    aws_secret_access_key=AWS_SECRET_ACCESS_KEY,
    aws_session_token=AWS_SESSION_TOKEN,
)

textract = session.client("textract", region_name = AWS_DEFAULT_REGION )
s3 = session.client("s3", region_name = AWS_DEFAULT_REGION)
```

Upload file if file does not exist.

```python
filename = 'dsv'
upload_prefix = 'raw-reports/'
file_ref_raw_file = upload_prefix + filename + '.pdf'
local_file_ref = file_ref_raw_file

try:
    s3.get_object(
        Bucket= BUCKET_NAME,
        Key= file_ref_raw_file
    )
    print("File already uploaded to S3")
except botocore.exceptions.ClientError as e:
    if e.response['Error']['Code'] == "NoSuchKey":
        print("Uploading to S3")
        with open(local_file_ref, "rb") as f:
            s3.upload_fileobj(f, BUCKET_NAME, file_ref_raw_file)
        print("Uploaded to S3")
    if e.response['Error']['Code'] == "ExpiredToken":
        print("Expired Token")


print("File is at: ", file_ref_raw_file);
```

Function to textract file asynchronously and fetch block response

```python
def textract_file(file_ref):
    doc_textracted_async = textract.start_document_analysis(
        DocumentLocation={
            'S3Object':{
                'Bucket': BUCKET_NAME,
                'Name': file_ref
            }
        },
        FeatureTypes=['TABLES']
    )
    job_id = doc_textracted_async["JobId"]
    all_blocks = []
    print("getting full_json")
    response = get_full_json(job_id, Textract_API.ANALYZE, textract)
    return response
```

Get textracted file if exist from S3, else textract and upload

```python
upload_textracted_prefix = 'textracted-reports-values/'
file_ref_tetracted = upload_textracted_prefix + filename + '.json'

document_json = {}

try:
    document_json = s3.get_object(
        Bucket= BUCKET_NAME,
        Key= file_ref_tetracted
    )
    document_json = json.loads(document_json["Body"].read().decode('UTF-8'))
    print("File already textracted")
    
except botocore.exceptions.ClientError as e:
    if e.response['Error']['Code'] == "NoSuchKey":
        print("Textracting.....")
        document_json = textract_file(file_ref_raw_file)
        print("Textracted")

        json_file = json.dumps(document_json).encode('UTF-8')

        s3.put_object(
            Body = json_file,
            Bucket = BUCKET_NAME,
            Key = file_ref_tetracted
        )
        print("File uploaded at: ", file_ref_tetracted)
all_blocks = document_json["Blocks"]
```

```python
print(get_string(textract_json=document_json, output_type=[Textract_Pretty_Print.TABLES]))
```

```python
## from trp import Document
from array import *
import numpy as np

doc = Document(document_json)

table_num = 0

all_array = []


# Iterate over elements in the document
for p, page in enumerate(doc.pages):
#     print("Page number: {}".format(p+1))
    
    # Print tables
    for t, table in enumerate(page.tables):
        table_array = []
#         print("Table number: {}".format(t+1))

        for r, row in enumerate(table.rows):
            row_array = []
            for c, cell in enumerate(row.cells):
                row_array.append(cell.text.strip())
#                 print("Table[{}][{}] = {}-{}".format(r, c, cell.text, cell.confidence))
            table_array.append(row_array)
        all_array.append(table_array)
#         print("\n")
        
print("Number of tables", len(all_array))
```

```python
import re

def is_header_years(header_array):
    is_years = True
    for idx, i in enumerate(header_array):
        if idx>0:
            if not re.match(r"(19|20|21)[0-9]{2}", i) or len(i)!=4:
                is_years = False
                break

    return is_years


# format year to be in 1st row

tables_with_year = []
tables_without_year = []


for idx, table in enumerate(all_array):
    suspect_header = table[0]
    if not is_header_years(suspect_header):
        table = np.transpose(np.array(table))
        suspect_header = table[0]
        if not is_header_years(suspect_header):
            tables_without_year.append(table)
            continue
    
    table[0][0] = "Year"
    tables_with_year.append(table)
    
```
