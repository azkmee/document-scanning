# Textract function
Function to textract file asynchronously and fetch block response. Although the function to start analysis is done asynchronously, getting full json requires waiting till the job is completed.

For future use in the cloud, a notification can be set up to notify completion of extraction

```python
def textract_file(file_ref):
    doc_textracted_async = textract.start_document_analysis(
        DocumentLocation={
            'S3Object':{
                'Bucket': BUCKET_NAME,
                'Name': file_ref
            }
        },
        FeatureTypes=['TABLES']
    )
    job_id = doc_textracted_async["JobId"]
    all_blocks = []
    print("getting full_json")
    response = get_full_json(job_id, Textract_API.ANALYZE, textract)
    return response
```

# Extract text from merge cells
AWS textract returns us merge cells detected in the tables. We would replicate the entries of all merge cells into the normal cells as we wont be able to process merge cells here

```python
# i want to go into a merge cell and replicate for the space
def extract_child_ids_if_present(block):
    
    try:       
        return block["Relationships"][0]["Ids"]
    except:
        return None

# get text of cell
def get_text_from_cell_ids(cell_ids, all_blocks_dict_by_id):
    word_ids = []
    
    for cell_id in cell_ids:
        child_id = extract_child_ids_if_present(all_blocks_dict_by_id[cell_id])
        if child_id != None:
            word_ids += child_id
            
    word_arr = [all_blocks_dict_by_id[word_id]["Text"] for word_id in word_ids]
    
    return (' ').join(word_arr)

# get text, row idx, row span, col idx, col span
def get_details_from_merge_cell(merge_block_id, all_blocks_dict_by_id):

    merge_block = all_blocks_dict_by_id[merge_block_id]
    
    #find all child id
    child_ids = merge_block["Relationships"][0]["Ids"]
    text = get_text_from_cell_ids(child_ids, all_blocks_dict_by_id)
    
#     return{
#         'Text': text,
#         'RowIndex': merge_block["RowIndex"],
#         'ColumnIndex': merge_block["ColumnIndex"],
#         'RowSpan': merge_block["RowSpan"],
#         'ColumnSpan': merge_block["ColumnSpan"],
#     }
    return [text, merge_block["RowIndex"], merge_block["ColumnIndex"], merge_block["RowSpan"], merge_block["ColumnSpan"]]
```

```python
table_blocks = [block for block in all_blocks if block["BlockType"] == "TABLE"]

all_table_arrays = []

# for all tables return by textract:
for idx, table_block in enumerate(table_blocks):
    relations = table_block['Relationships']
    child_ids = []
    merged_cells_ids = []

    # check for child and merge cell relationship
    for relation in relations:
        if relation["Type"] == "CHILD":
            child_ids = relation["Ids"]
        elif relation["Type"] == "MERGED_CELL":
            merged_cells_ids = relation["Ids"]

    current_row = 1
    table_array = []
    row_array = []

    
    # compile cells into array
    for cell_id in child_ids:
        blocks = all_blocks_dict_by_id[cell_id]

        text = get_text_from_cell_ids([cell_id],all_blocks_dict_by_id)
        row_index =  blocks["RowIndex"]
        column_index = blocks["ColumnIndex"]

        if current_row == row_index:
            row_array.append(text)
        else:
            table_array.append(row_array)
            row_array = [text]
            current_row = row_index

    table_array.append(row_array)

    # refactor cells that has merged values (currently only for row)
    for merged_cell_id in merged_cells_ids:
        text, row_idx, col_idx, row_span, col_span  = get_details_from_merge_cell(merged_cell_id, all_blocks_dict_by_id)
#         print(text, row_idx, col_idx, row_span, col_span)

        row_idx_start = row_idx-1
        row_idx_end = row_idx-1 + row_span
        for row_idx in range(row_idx_start, row_idx_end):
            table_array[row_idx][col_idx-1] = text
            
    # combine all table        
    all_table_arrays.append(table_array)
```

# Key Word Mapping
List down keywords to extract from the table. All keys in list have to exist.

```python
year = 2021
sample_keywords = {
    "Total Scope 1 GHG emissions ('000 tonnes)":['Total', 'Scope 1'],
    "Total Scope 2 GHG emissions ('000 tonnes)":['Total', 'Scope 2'],
    "Total Scope 3 GHG emissions ('000 tonnes)":['Total', 'Scope 3'],
    "Total electricity consumption ('000 gigajoule)":['Total', 'energy','consumption'],
    "Total water consumption ('000 m3)":['Total', 'water', 'consumption'],
#     "Total waste generated (tonnes)":[],
#     "Total Hazardous waste (tonnes)":[],
#     "Total Non-Hazardous waste (tonnes)":[],
    "Current gender diversity (female/male) (%)":['gender', 'diversity', 'total'],
    "Current gender diversity in management (female/male) (%)":['gender', 'diversity', 'management'],
    "Number of severe injury":["fatalities", "absolute", "number"]
#     "%of employees by age group (%)":[],
#     "Average training hours per employee (hours)":[],
#     "Turnover rate (%)":[],
#     "%women on the board (%)":[],
#     "%women in Management (%)":[],
#     "Number of corruption case":[]    
}
```

```python
import re

def is_year(number):
    if re.match(r"(19|20|21)[0-9]{2}", number) and len(number)==4:
        return True
    else:
        False
    
# returns False if not found
# return location of year and key location if found
def loc_for_year_header(year, table_array):
    year_loc_row = {}
    year_loc_col = {}
    
    num_of_rows = len(table_array)
    num_of_col = len(table_array[0])
    
    array_year = np.zeros((num_of_rows, num_of_col))
    for row_id, row in enumerate(table_array):
        
        for col_id, elem in enumerate(row):
            if is_year(elem):
                array_year[row_id][col_id] = 1
                if str(elem) == str(year):
                    # in case we have elements with year format above
                    try:
                        year_loc_row[row_id].extend([col_id])
                        year_loc_col[col_id].extend([row_id])
                    except:
                        year_loc_row[row_id] = [col_id]
                        year_loc_col[col_id] = [row_id]
                                            
    # in future, need to factor in if year is not in first col or first row
    row_sum = np.sum(array_year, axis = 0)[0] #row sum
    col_sum = np.sum(array_year, axis = 1)[0] #col sum
    
    #only support year in 1 year or 1 col
    if row_sum > col_sum: #year is in first col
        row_idxs = year_loc_col[0]
        if len(row_idxs) != 1:
            return False
        else:
            return ((row_idxs[0], 0), "row")
    else: #year is in first row
        col_idxs = year_loc_row[0]
        if len(col_idxs) != 1:
            return False
        else:
            return ((0, col_idxs[0]), "col")
```

```python
# year loc used to know where to start extracting data
year_loc, key_loc = loc_for_year_header(year, arr_np)
# print(year_loc, key_loc)

header_mapping = {}
filtered_table_array = []

for table_arr in all_table_arrays:
    if key_loc == 'col':
        for row in table_arr:
            table_key = row[0].lower()
            for key in sample_keywords:
    #             print("key", key)
                is_table_key = 1
                for values in sample_keywords[key]:
#                     print("values", values)
                    if values.lower() not in table_key:
                        is_table_key = 0
    #                     print("values not in /", table_key)
                        break
    #                 print("in value for loop")
#                     print("values", values, table_key)
                if is_table_key:
                    if row[year_loc[1]] != '':
                        header_mapping[key]=table_key
                        filtered_table_array.append([key] + row[1:])
#                     print(key, row[year_loc[1]])
                    
    #             print("in key for loop")


            
```
